```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt

women_degrees = pd.read_csv('percent-bachelors-degrees-women-usa.csv')
cb_dark_blue = (0/255,107/255,164/255)
cb_orange = (255/255, 128/255, 14/255)
stem_cats = ['Psychology', 'Biology', 'Math and Statistics', 'Physical Sciences', 'Computer Science', 'Engineering']
lib_arts_cats = ['Foreign Languages', 'English', 'Communications and Journalism', 'Art and Performance', 'Social Sciences and History']
other_cats = ['Health Professions', 'Public Administration', 'Education', 'Agriculture','Business', 'Architecture']

fig, ax = plt.subplots(6, 3, figsize = (24, 24))
fig.tight_layout()

for sp in range(0,len(stem_cats)):
    ax[sp, 0].plot(women_degrees['Year'], women_degrees[stem_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax[sp, 0].plot(women_degrees['Year'], 100-women_degrees[stem_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax[sp, 0].spines["right"].set_visible(False)    
    ax[sp, 0].spines["left"].set_visible(False)
    ax[sp, 0].spines["top"].set_visible(False)    
    ax[sp, 0].spines["bottom"].set_visible(False)
    ax[sp, 0].set_xlim(1968, 2011)
    ax[sp, 0].set_ylim(0,100)
    ax[sp, 0].set_title(stem_cats[sp], fontdict = {"fontsize"})
    ax[sp, 0].tick_params(bottom="off", top="off", left="off", right="off", labelbottom = False)
    ax[sp, 0].set_yticks([0,100])
    ax[sp, 0].axhline(50, c=(171/255, 171/255, 171/255), alpha = 0.3)
    if sp == 0:
        ax[sp, 0].text(2005, 87, 'Men')
        ax[sp, 0].text(2002, 8, 'Women')
    elif sp == len(stem_cats)-1:
        ax[sp, 0].text(2005, 62, 'Men')
        ax[sp, 0].text(2001, 35, 'Women')
        ax[sp, 0].tick_params(labelbottom = True)
        
for sp in range(0, len(lib_arts_cats)):
    ax[sp, 1].plot(women_degrees['Year'], women_degrees[lib_arts_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax[sp, 1].plot(women_degrees['Year'], 100-women_degrees[lib_arts_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax[sp, 1].spines["right"].set_visible(False)    
    ax[sp, 1].spines["left"].set_visible(False)
    ax[sp, 1].spines["top"].set_visible(False)    
    ax[sp, 1].spines["bottom"].set_visible(False)
    ax[sp, 1].set_xlim(1968, 2011)
    ax[sp, 1].set_ylim(0,100)
    ax[sp, 1].set_title(lib_arts_cats[sp], font_size = 20)
    ax[sp, 1].tick_params(bottom="off", top="off", left="off", right="off", labelbottom = False)
    ax[sp, 1].set_yticks([0,100])
    ax[sp, 1].axhline(50, c=(171/255, 171/255, 171/255), alpha = 0.3)
    if sp == 0:
        ax[sp, 1].text(2005, 87, 'Men')
        ax[sp, 1].text(2002, 8, 'Women')
    elif sp == len(lib_arts_cats)-1:
        ax[sp, 1].tick_params(labelbottom = True)
        ax[sp+1, 1].set_visible(False)
        
for sp in range(0, len(other_cats)):
    ax[sp, 2].plot(women_degrees['Year'], women_degrees[other_cats[sp]], c=cb_dark_blue, label='Women', linewidth=3)
    ax[sp, 2].plot(women_degrees['Year'], 100-women_degrees[other_cats[sp]], c=cb_orange, label='Men', linewidth=3)
    ax[sp, 2].spines["right"].set_visible(False)    
    ax[sp, 2].spines["left"].set_visible(False)
    ax[sp, 2].spines["top"].set_visible(False)    
    ax[sp, 2].spines["bottom"].set_visible(False)
    ax[sp, 2].set_xlim(1968, 2011)
    ax[sp, 2].set_ylim(0,100)
    ax[sp, 2].set_title(other_cats[sp], font_size = 20)
    ax[sp, 2].tick_params(bottom="off", top="off", left="off", right="off", labelbottom = False)
    ax[sp, 2].set_yticks([0,100])
    ax[sp, 2].axhline(50, c=(171/255, 171/255, 171/255), alpha = 0.3)
    if sp == 0:
        ax[sp, 2].text(2005, 87, 'Men')
        ax[sp, 2].text(2002, 8, 'Women')
    elif sp == len(other_cats)-1:
        ax[sp, 2].text(2005, 62, 'Men')
        ax[sp, 2].text(2001, 35, 'Women')
        ax[sp, 2].tick_params(labelbottom = True)
plt.savefig("gender_degrees.png")
```
